```python
# Read the sensor rays

import numpy as np

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import mpl_toolkits.mplot3d as plt3d
```

```python
#unit vectors
x = np.array([1,0,0,1])
x = x.transpose()
print(x)
y = np.array([0,1,0,1])
y = y.transpose()
z = np.array([0,0,1,1])
z = z.transpose()
origin = np.array([0,0,0,1])

scale = [0.1,0.1,0.1,1]
x = scale * x
y = scale * y
z = scale * z
```

```python
file = "/home/irving/projects/partialmodel/data_example/FreeFlyer_kriegel_blensor/config/sensorRays.dat"
rays = np.loadtxt(file)
print(rays)
```

```python


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

rays_t = np.transpose(rays)
ax.scatter(rays_t[0],rays_t[1],rays_t[2])
ax.scatter(0,0,0)

# draw axis
xxs = origin[0], x[0]
xys = origin[1], x[1]
xzs = origin[2], x[2]
    
line = plt3d.art3d.Line3D(xxs, xys, xzs, color="red")
ax.add_line(line)
    
yxs = origin[0], y[0]
yys = origin[1], y[1]
yzs = origin[2], y[2]
    
line = plt3d.art3d.Line3D(yxs, yys, yzs, color="green")
ax.add_line(line)
    
zxs = origin[0], z[0]
zys = origin[1], z[1]
zzs = origin[2], z[2]
    
line = plt3d.art3d.Line3D(zxs, zys, zzs, color="blue")
ax.add_line(line)
    

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

```python
file = "/home/irving/projects/partialmodel/data_example/FreeFlyer_kriegel_blensor/data/candidate_views.txt"

infile = open(file, 'r')
# read next line:
n_views = int(infile.readline())

matrices = []

for i in range(n_views):
    lv = int(infile.readline())
    #print(lv)
    lw = int(infile.readline())
    #print(lw)
    words = infile.readline().split()
    #print(words)
    words = infile.readline().split()
    #print(words)
    #pose = float()
    r1 = infile.readline().split()
    r1 = [float(j) for j in r1]
    #print(r1)
    r2 = infile.readline().split()
    r2 = [float(j) for j in r2]
    #print(r2)
    r3 = infile.readline().split()
    r3 = [float(j) for j in r3]
    #print(r3)
    r4 = infile.readline().split()
    r4 = [float(j) for j in r4]
    #print(r4)
    
    m = np.array([r1,r2,r3,r4])
    #print(m)
    matrices.append(m)
    
```

```python
fig = plt.figure()
fig.set_size_inches(10,10)

ax = fig.add_subplot(111, projection='3d', aspect='equal', xlim = (-0.5,0.5), ylim = (-0.5,0.5), zlim = (0.5,1.5))
ax.view_init(azim=120)


for M in matrices:
    #plot the lines
    #rint(x)
    xp = np.dot(M,x)
    #rint(xp)
    yp = np.dot(M,y)
    zp = np.dot(M,z)
    
    origin_p = np.dot(M, origin)
    #rint(origin_p)
    
    xxs = origin_p[0], xp[0]
    xys = origin_p[1], xp[1]
    xzs = origin_p[2], xp[2]
    
    line = plt3d.art3d.Line3D(xxs, xys, xzs, color="red")
    ax.add_line(line)
    
    yxs = origin_p[0], yp[0]
    yys = origin_p[1], yp[1]
    yzs = origin_p[2], yp[2]
    
    line = plt3d.art3d.Line3D(yxs, yys, yzs, color="green")
    ax.add_line(line)
    
    zxs = origin_p[0], zp[0]
    zys = origin_p[1], zp[1]
    zzs = origin_p[2], zp[2]
    
    line = plt3d.art3d.Line3D(zxs, zys, zzs, color="blue")
    ax.add_line(line)
    
    #reak
    

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
    
#plt.plot()
```

```python
from random import randint

# Draw the sensor for a given view
fig = plt.figure()
fig.set_size_inches(10,10)

ax = fig.add_subplot(111, projection='3d', aspect='equal', xlim = (-0.5,0.5), ylim = (-0.5,0.5))
ax.view_init(azim=120)

i = randint(0, n_views-1);
print(i)
M = matrices[i]


#plot the lines
#rint(x)
xp = np.dot(M,x)
#rint(xp)
yp = np.dot(M,y)
zp = np.dot(M,z)
    
origin_p = np.dot(M, origin)
print(origin_p)
    
xxs = origin_p[0], xp[0]
xys = origin_p[1], xp[1]
xzs = origin_p[2], xp[2]
    
line = plt3d.art3d.Line3D(xxs, xys, xzs, color="red")
ax.add_line(line)
   
yxs = origin_p[0], yp[0]
yys = origin_p[1], yp[1]
yzs = origin_p[2], yp[2]
    
line = plt3d.art3d.Line3D(yxs, yys, yzs, color="green")
ax.add_line(line)
    
zxs = origin_p[0], zp[0]
zys = origin_p[1], zp[1]
zzs = origin_p[2], zp[2]
    
line = plt3d.art3d.Line3D(zxs, zys, zzs, color="blue")
ax.add_line(line)

rotated_rays = []
# rotate the 
for r in rays:
    r_h = np.array([r[0], r[1], r[2], 1])
    r_p = np.dot(M, r_h.transpose())
    rotated_rays.append([r_p[0],r_p[1],r_p[2]])
    
rays_t = np.transpose(rotated_rays)
ax.scatter(rays_t[0],rays_t[1],rays_t[2])
ax.scatter(0,0,0)

```
